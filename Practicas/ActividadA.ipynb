```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos las matrices E y E1 con las nuevas coordenadas
E = np.array([
    (0, 0, 1), (4, 11, 1), (8, 11, 1), (12, 0, 1), (10, 0, 1), (9, 4, 1), (3, 4, 1), (2, 0, 1), (0, 0, 1)
])

E1 = np.array([
    (5, 9, 1), (7, 9, 1), (8, 6, 1), (4, 6, 1), (5, 9, 1)
])

# Definir matrices de transformación
Ic = np.eye(3)

# Matriz de Reflexión con respecto al eje x
Refx = np.array([[1., 0, 0], [0, -1., 0], [0., 0., 1.]])

# Matriz de Reflexión con respecto al eje y
Refy = np.array([[-1., 0, 0], [0, 1., 0], [0., 0., 1.]])

# Matriz de Rotación
theta = np.pi / 3  # Ángulo de rotación deseado
R = np.array([
    [np.cos(theta), np.sin(theta), 0.],
    [-np.sin(theta), np.cos(theta), 0.],
    [0., 0., 1.]
])

# Matriz de cambio de escala
s = 2  # Escalar
S = np.array([
    [s, 0, 0.],
    [0., s, 0.],
    [0., 0., 1.]
])

# Matriz de deformación horizontal/vertical
h = -1
v = 2
D = np.array([
    [1., h, 0.],
    [v, 1., 0.],
    [0., 0., 1.]
])

# Matriz de traslación
tx = -3
ty = -5
T = np.array([
    [1., 0, tx],
    [0, 1., ty],
    [0., 0., 1.]
])

# Graficar transformaciones aplicadas a ambas matrices
def graficar_transformaciones_juntas(E, E1, transformacion, titulo):
    Ex, Ey = [], []
    Ex1, Ey1 = [], []

    # Aplicar la transformación a E
    for row in E:
        output_row = transformacion.dot(row)
        x, y, _ = output_row
        Ex.append(x)
        Ey.append(y)

    # Aplicar la transformación a E1
    for row in E1:
        output_row = transformacion.dot(row)
        x, y, _ = output_row
        Ex1.append(x)
        Ey1.append(y)

    # Graficar las matrices originales y transformadas
    plt.figure(figsize=(8, 8))
    plt.plot(E[:, 0], E[:, 1], color="blue", label="Original E")
    plt.plot(E1[:, 0], E1[:, 1], color="blue", label="Original E1")
    plt.plot(Ex, Ey, color="red", label=f"Transformada E ({titulo})")
    plt.plot(Ex1, Ey1, color="red", label=f"Transformada E1 ({titulo})")
    ax = plt.gca()
    ax.set_xticks(np.arange(-15, 20, 5))
    ax.set_yticks(np.arange(-15, 20, 5))
    ax.set_aspect('equal', adjustable='box')
    plt.title(f"Transformación: {titulo}")
    plt.grid()
    plt.legend()
    plt.show()

# Aplicar y graficar cada transformación
graficar_transformaciones_juntas(E, E1, Refx, "Reflexión en X")
graficar_transformaciones_juntas(E, E1, Refy, "Reflexión en Y")
graficar_transformaciones_juntas(E, E1, R, "Rotación π/3")
graficar_transformaciones_juntas(E, E1, S, "Escalado por 2")
graficar_transformaciones_juntas(E, E1, D, "Deformación (h=-1, v=2)")
graficar_transformaciones_juntas(E, E1, T, "Traslación (-3, -5)")
```
