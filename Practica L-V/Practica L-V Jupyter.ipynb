### LIBRERIAS

```python
import matplotlib.pyplot as plt
from random import *
from numpy import *
import sys
```

#### CÃ“DIGO ORIGINAL

```python
a = 0.7; b = 0.5; c = 0.3; e = 0.2
dt = 0.001; max_time = 100
t = 0; x = 1.0; y = 0.5
t_list = []; x_list = []; y_list = []
t_list.append(t); x_list.append(x); y_list.append(y)
while t < max_time:
    t = t + dt
    x = x + (a*x - b*x*y)*dt
    y = y + (-c*y + e*x*y)*dt

    t_list.append(t)
    x_list.append(x)
    y_list.append(y)

p = plt.plot(t_list, x_list, 'r', t_list, y_list, 'g', linewidth = 2) 
```

En el anterior grÃ¡fico podemos observar que el aumento de la presencia de las presas
es mayor que el de los depredadores dado que el aumento natural de las presas es de
(0.7) y la destrucciÃ³n por depredador es de (0.5), con ello podemos ver que se
reproducen mÃ¡s rÃ¡pido las presas que la muerte por depredador, al igual vemos que
la muerte de los depredadores es mayor con (0.3) que el aumento de los mismo por
alimento de las presas con (0.2)
